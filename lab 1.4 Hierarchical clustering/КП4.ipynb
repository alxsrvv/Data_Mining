#Ієрархічна кластеризація


**Ієрархічна кластеризація** (також графові алгоритми кластеризації і ієрархічний кластерний аналіз) - сукупність алгоритмів упорядкування даних, спрямованих на створення ієрархії (дерева) вкладених кластерів. В результаті даного алгоритму отримують дендрограму. 

Під дендрограмою зазвичай розуміють дерево, побудоване за значеннями матриці близькості. Дендрограмма дозволяє зобразити взаємні зв'язки між об'єктами з заданої множини. 

Виділяють два класи методів ієрархічної кластеризації:

*Агломеративні методи* (англ. Agglomerative): нові кластери створюються шляхом об'єднання дрібніших кластерів і, таким чином, дерево створюється від листя до стовбура;

*Дивізівні або дивізійні методи* (англ. Divisive): нові кластери створюються шляхом ділення більших кластерів на більш дрібні і, таким чином, дерево створюється від стовбура до листя.

Алгоритми ієрархічної кластеризації припускають, що аналізована множина об'єктів характеризується певним ступенем зв'язності. За кількістю ознак іноді виділяють монотетічені та політетічні методи класифікації. Як і більшість візуальних способів подання залежностей графи швидко втрачають наочність при збільшенні числа кластерів. Найчастіше використовуються агломеративні методи.

3. Об'єднаємо дані об'єкти в один клас, визначивши центр даного кластеру як середньоарифметичне по всім параметрам.

4. Повторимо крок 2, з врахуванням того, що об'єкти 8-10 предавляють собою один клас.

## Завдання
1. Реалізувати ієрархічну кластеризацію.
2. Вивести таблицю об'єднання кластерів та дендрограму (побудова графіку не обов'язкова, але буде перевагою)
3. З набору даних в КП2 відібрати 20 випадкових об'єктів та провести кластерний аналіз.
4. Зробити висновки з обов'язковим вказанням оптимальної кількості кластерів.
