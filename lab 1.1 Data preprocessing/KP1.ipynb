```python
from sklearn.preprocessing import Binarizer
import pandas as pd
import numpy as np

#url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/titanic.csv'
#df1 = pd.read_csv(url)
df1 = pd.read_csv('titanic.csv')
#pd.set_option("display.max_rows", None, "display.max_columns", None)
df1
```

```python
def indicator_average(data, column_to_change):
    data.insert(data.columns.get_loc(column_to_change)+1, f'{column_to_change}_was_n\\a', data[column_to_change].isna()) 
    data[column_to_change] = data[column_to_change].fillna(data[column_to_change].mean())
indicator_average(df1, 'Age')
indicator_average(df1, 'Fare')

df1
```

```python
def z_score(data, column_to_change):
    data[column_to_change] = (data[column_to_change] - data[column_to_change].mean())/data[column_to_change].std(ddof=0)
def min_max(data, column_to_change):
    data[column_to_change] = (data[column_to_change]-data[column_to_change].min())/(data[column_to_change].max()-data[column_to_change].min())
#z_score(df1, 'Age')
min_max(df1, 'Fare')

df1
```

```python
# параметры: (таблица, столбец_для_бинаризации, пороговое_значение)
def binarization(data, column_to_change, trigger):
    data[column_to_change] = np.where(data[column_to_change] > trigger, 1, 0)
def bool_to_num(data, column_to_change):
    data[column_to_change] = np.where(data[column_to_change] == True, 1, 0)
    
#binarization(df1, 'Fare', 15)
binarization(df1, 'Age', 18)
bool_to_num(df1, 'Age_was_n\\a')
bool_to_num(df1, 'Fare_was_n\\a')

df1
```

```python
def text_coding(data, column_to_change):
    data[column_to_change], uniques = pd.factorize(data[column_to_change])
    print(uniques) # вывод значений кодирования
text_coding(df1, 'Embarked')
text_coding(df1, 'Sex')
#text_coding(df1, 'class')
#text_coding(df1, 'who')
#text_coding(df1, 'deck')
#text_coding(df1, 'embark_town')

df1
```
