## Теоретичні відомості

Сімейство алгоритмів КРАБ засновано на формуванні незамкнутого зв'язного графа і проведенні розбиття множини вихідних об'єктів з його допомогою. Ці алгоритми дозволяють будувати класи довільної форми і розмірів.

Природно, що людина прагне використовувати деяку форму близькості (ρ) точок і вважає, що класифікація тим краще, чим менше відстань між точками одного і того ж класу. Крім того, людина впевненіше виділяє класи, якщо "згустки" точок більш віддалені один від одного, так що вводиться міра віддаленості (d). За інших рівних умов людину більше влаштовує, коли розподіл точок по класах приблизно рівномірний (h).  Психологічні експерименти показали, що людина не завжди об'єднує точки в один клас за принципом «ближній до ближнього». Так, для наступного зображення п'ята за рахунком точка ближче до четвертої, але людина проведе кордон класів між четвертою та п'ятою точками. Він звертає увагу на зміни щільності точок (λ). якщо підібрати відповідні заходи для (ρ), (d), (h) і (λ), то можна домогтися збігу результатів автоматичного і ручного груповання.

Через КНП можна визначити і *міру локальної неоднорідності відстаней* між точками. Для цього позначимо довжину деякого ребра $a_{i}$, а довжину найменшого примикаючого до нього ребра через $β_{min}$. Загальний критерій якості в алгоритмі КРАБ сформульований так:

image.png

1. Вибір початкової точки з числа об`єктів у вибірці.
2. Знаходження відстані від даної точки до всіх об'єктів вибірки.
3. Знаходження пар точок з мінімальним значенням відстані між ними. Ці точки з'єднуються ребром графа. 
4. Розрахувати відстань від останньої з приєднаних точок до точок, що ще не є частиною графа.
5. Повторити пункти 3-4.

Ця процедура повторюється до тих пор, коли всі об'єкти не будуть з'єднаними ребрами графу. Так граф не буде мати петель, а сумарна довжина всіх його ребер буде мінімальною. Граф, що має такі властивості, називається найкоротшим незамкнутим шляхом (ННШ).

Якщо розрізати одне ребро, то вийде два класи, якщо розрізати (К-1) ребро КНП, ми отримаємо К класів точок.

image.png

Якщо задано число класів (K), на яке необхідно провести розбиття, то шляхом перебору вибираються (K-1) ребер, проведення меж за якими дасть максимальне значення функціонала F.
Якщо досліднику не важлива рівномірність класів за кількістю об'єктів, то використовується модифікація F.
